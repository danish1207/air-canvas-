```python
import cv2
import numpy as np
# capture the video
cap2=cv2.VideoCapture(0)
cap2.set(3,2980)
cap2.set(4,2080)
cap2.set(10,100)
#color found out using video color detection(we can add more color here)
mycolor=[[30,72,81,72,249,255]]

colorvalues=[[0,255,255]]

mypoints=[]   #x,y,colorid
def findcolor(img,mycolor,colorvalues):

    imghsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
    count=0
    newpoints=[]
    for color in mycolor:
        lower = np.array(color[0:3])
        upper = np.array(color[3:6])
        mask = cv2.inRange(imghsv, lower, upper)
        #cv2.imshow(str(color[0]), mask)
        x,y=getcontours(mask)
        cv2.circle(imgresult,(x,y),1,colorvalues[count],cv2.FILLED)
        if x!=0 and y!=0:
            newpoints.append([x,y,count])
        count+=1
    return newpoints

#creating boundaries around our pens
def getcontours(img):
    countours, hierachy = cv2.findContours(img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    x,y,w,z=0,0,0,0
    for i in countours:
        area = cv2.contourArea(i)
        if area > 500:
            #cv2.drawContours(imgresult, i, -1, (255, 0, 0), 3)
            # perimeter of objects
            peri = cv2.arcLength(i, True)

            # corner pts of all object
            arroxcorner = cv2.approxPolyDP(i, 0.09 * peri, True)
            # creating a rect around the object
            x, y, w, h = cv2.boundingRect(arroxcorner)
    return x+w//2,y #returning the tip


def drawoncanvus(mypoints,colorvalues):
    for pt in mypoints:
        cv2.circle(imgresult, (pt[0],pt[1]), 20, colorvalues[pt[2]], cv2.FILLED)
while True:
    success,img=cap2.read()
    imgresult=img.copy()
    newpoints=findcolor(img,mycolor,colorvalues)
    if len(newpoints)!=0:
        for newp in newpoints:
            mypoints.append(newp)
    if len(mypoints)!=0:
        drawoncanvus(mypoints,colorvalues)

    cv2.imshow("otp",imgresult)
    if cv2.waitKey(1)&0XFF==ord('q'):
        break
cv2.destroyAllWindows()
cap2.release()
```
